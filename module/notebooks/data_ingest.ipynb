```python
import yfinance as yf
import pandas as pd
from tqdm import tqdm
from pathlib import Path

# Load config (if using Hydra, otherwise hardcode)
DATA_PATH = Path("../data/raw")
DATA_PATH.mkdir(parents=True, exist_ok=True)

# Example ticker list
tickers = ["AAPL", "MSFT", "GOOG", "AMZN", "TSLA", "NVDA", "META", "NFLX"]  # extend to 100–200

all_data = []
for t in tqdm(tickers):
    df = yf.download(t, start="2018-01-01", end="2025-01-01", progress=False)
    df["Ticker"] = t
    all_data.append(df)

full_df = pd.concat(all_data)
full_df.to_parquet(DATA_PATH / "ohlcv.parquet")

print("✅ Saved OHLCV data to", DATA_PATH / "ohlcv.parquet")
```
